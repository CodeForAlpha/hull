```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root
```

```python
# 투자 정보
capital = 5800
current_price = 29

# 투자 전략
## 1. 콜 옵션 매수
strike_price = 30
premium = 2.9
options_bought = int(capital / premium)

## 2. 주식 직접 매수
stocks_bought = int(capital / current_price)
```

```python
# P&L 함수 정의
## 1. 콜 옵션 매수의 P&L
def option_pnl(stock_price_t):
    # 옵션의 내재가치
    current_option_value = np.maximum(stock_price_t - strike_price, 0) * options_bought
    # 프리미엄 지급비용
    option_premium = options_bought * premium
    return current_option_value - option_premium

## 2. 주식 매수의 P&L
def stock_pnl(stock_price_t):
    return (stock_price_t - current_price) * stocks_bought
```

```python
s_t = np.arange(0, 70, 0.1)
pnl_option = option_pnl(s_t)
pnl_stock = stock_pnl(s_t)
```

```python
# 교차점 표시
def error_function(s_t):
    s_t = s_t[0]
    return option_pnl(s_t) - stock_pnl(s_t)

result = root(error_function, x0=[30.0])
intersection_stock_price = result.x[0]
intersection_pnl = option_pnl(intersection_stock_price)




plt.figure(figsize=(10, 6))

# line graphs
plt.plot(s_t, pnl_option, label="option", color="blue")
plt.plot(s_t, pnl_stock, label="stock", color="green")

# intersection
plt.plot(intersection_stock_price, intersection_pnl, 'o', color='purple', markersize=8, 
             label=f":(${intersection_stock_price:.2f}, ${intersection_pnl:.2f})")

# x-axis
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(current_price, color='black', linestyle=':', linewidth=0.8, label=f"current price (${current_price})")
plt.axvline(strike_price, color='black', linestyle=':', linewidth=0.8, label=f"strike price (${strike_price})")

plt.title("P&L Comparison of Investment Strategies")
plt.xlabel("Expiration price")
plt.ylabel("Total P&L")
plt.xlim(s_t.min(), s_t.max())
plt.ylim(pnl_option.min() - 500, pnl_stock.max() + 500)
plt.grid(True, linestyle='dotted', alpha=0.6)
plt.legend()
plt.show()
```
