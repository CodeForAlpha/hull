```python
import random
import numpy

p_h = 0.99
p_t = 0.01

def throw_coin():
    if random.random() <= p_h:
        return 1
    else:
        return -100

def simulate(num_throws):
    total_gain = 0
    for _ in range(num_throws):
        total_gain += throw_coin()
    return total_gain

results = []
for _ in range(100000):
    result = simulate(3500)
    results.append(result)


print(numpy.mean(results))
print(numpy.var(results))

for _ in range(100000):
    result = simulate(3500)
    results.append(result)


print(numpy.mean(results))
print(numpy.var(results))
```
